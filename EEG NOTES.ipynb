# EEG

EEG is a test that measures <u>electrical activity</u> of the brain. The test uses small, metal discs called electrodes that attach to the scalp. Brain cells communicate via electrical impulses, and this activity shows up as wavy lines on an EEG recording. Brain cells are active all the time, even during sleep.

EEG records the __summed electrical activity__ of thousands of neurons, primarily postsynaptic potentials in the brain's cortical regions. The activity is represented as waveforms with variations in amplitude (strength) and frequency (speed). EEG has excellent temporal resolution, capturing neural activity in milliseconds. This makes it ideal for studying fast brain processes like perception, attention, and decision-making.

Neurons in the brain communicate using electrical signals. When groups of neurons fire together, they generate electrical fields detectable by electrodes.

$$
\begin{array}{|c|c|c|}
\hline
\textbf{Frequency Band} & \textbf{Frequency (Hz)} & \textbf{Associated Activity} \\ \hline
Delta & 0.5\text{ - }4 \, \text{Hz} & \text{ Deep sleep, slow-wave sleep } \\ \hline
Theta & 4\text{ - }8 \, \text{Hz} & \text{ Drowsiness, meditation, light sleep }\\ \hline
Alpha & 8\text{ - }13 \, \text{Hz} & \text{ Relaxation, calm wakefulness }\\ \hline
Beta & 13\text{ - }30 \, \text{Hz} & \text{ Active thinking, problem-solving, alertness }\\ \hline
Gamma & 30+\text{Hz} & \text{ High-level cognition, memory, and perception }\\ \hline
\end{array}
$$


## CHANNELS

In EEG (Electroencephalography), channels refer to the electrodes used to record electrical activity from different regions of the brain. Each channel represents the electrical potential difference (voltage) measured between a specific electrode and a reference electrode

### Channel count

EEG systems can have varying numbers of channels, commonly ranging from 8 to 256 or more. The number of channels determines the spatial resolution of the EEG recording.

Typical setups include:
- __8-16 channels__: For basic clinical applications.
- __32-64 channels__: Common in research.
- __128-256 channels__: High-resolution systems for advanced studies.

$$
\begin{array}{|c|c|c|}
\hline
\textbf{System} & \textbf{Number of Channels} & \textbf{Use Case} \\ \hline
10\text{-}20 \, \text{System} & 19 & \text{ Standard clinical and research use }\\ \hline
10\text{-}10 \, \text{System} & 64 & \text{ Higher density for advanced research }\\ \hline
10\text{-}5 \, \text{System} & 128 & \text{ High spatial resolution studies }\\ \hline
\text{High-Density EEG} & 256+ & \text{ Source localization and connectivity analysis }\\ \hline
\end{array}
$$
