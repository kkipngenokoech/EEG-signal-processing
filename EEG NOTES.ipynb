# EEG

EEG is a test that measures <u>electrical activity</u> of the brain. The test uses small, metal discs called electrodes that attach to the scalp. Brain cells communicate via electrical impulses, and this activity shows up as wavy lines on an EEG recording. Brain cells are active all the time, even during sleep.

EEG records the __summed electrical activity__ of thousands of neurons, primarily postsynaptic potentials in the brain's cortical regions. The activity is represented as waveforms with variations in amplitude (strength) and frequency (speed). EEG has excellent temporal resolution, capturing neural activity in milliseconds. This makes it ideal for studying fast brain processes like perception, attention, and decision-making.

Neurons in the brain communicate using electrical signals. When groups of neurons fire together, they generate electrical fields detectable by electrodes.

## Definitions

__Spatial Resolution__: Spatial resolution in EEG refers to the ability to distinguish brain activity from different regions of the brain. Higher spatial resolution allows for more precise localization of brain activity, while lower resolution means signals from nearby regions may overlap. More electrodes improve spatial resolution, allowing for finer distinctions between brain areas.

## CHANNELS

In EEG (Electroencephalography), channels refer to the electrodes used to record electrical activity from different regions of the brain. Each channel represents the electrical potential difference (voltage) measured between a specific electrode and a reference electrode

### Channel count

EEG systems can have varying numbers of channels, commonly ranging from 8 to 256 or more. The number of channels determines the spatial resolution of the EEG recording.

Typical setups include:
- __8-16 channels__: For basic clinical applications.
- __32-64 channels__: Common in research.
- __128-256 channels__: High-resolution systems for advanced studies.

### EEG CHANNEL CONFIGURATIONS

This defines how electrodes are placed on the scalp and connected to a recording equipment. These configurations are critical for ensuring that the recorded data is meaningful and interpretable. The configurations are typically designed based on the application (clinical, research, or BCI) and the number of electrodes used.

The placement of EEG electrodes follows standardized systems, the most common being the 10-20 International System:

#### Electrode Labels:

- F: Frontal
- T: Temporal
- C: Central
- P: Parietal
- O: Occipital

__Even numbers__: Right side of the head (e.g., F4, T6).

__Odd numbers__: Left side of the head (e.g., F3, T5).

__Z__: Midline (e.g., Fz, Cz, Pz).

#### Distances:

Electrodes are spaced at 10% or 20% of the total head circumference, ensuring reproducibility across studies

An EEG (electroencephalogram) with 64 channels refers to a type of brain activity monitoring system that utilizes 64 electrodes or sensors placed on the scalp to record electrical activity generated by the brain. ### MONO-POLAR AND BI-POLAR EEG CHANNELS

The difference between monopolar and bipolar eeg recordings, lies in how electrodes are configured and how signals are referenced. 

In __monopolar EEG channels__, all electrodes are referred to as a single common reference electrode. This is commonly placed on the earbone or mastoid bone, usually at quite some distance from the remaining electrodes. An advantage of the monopolar channel is that it provides a more comprehensive brain activity measurement, but this comes with the risk of having more noise interfering with the data. 

Common Reference Options:
- __Earlobes__: Often considered electrically neutral.
- __Mastoids__: Bony area behind the ear.
- __Average Reference__: Mean of all electrode signals used as a reference.

In __bipolar EEG channels__, the signal is recorded as the difference in voltage between two adjacent electrodes. 



## REGIONS OF THE BRAIN

### Frontal Lobe

This is associated with decision making, motor control, problem solving and emotions. The frontal lobe has specialized subregions for these specific tasks (e.g., motor cortex for movement, prefrontal cortex for decision-making). Recording signals from electrodes positioned over specific regions (e.g., F3, F4, C3, C4 in EEG) helps isolate relevant activities.

We could also analyze the frequency domain to identify specific patterns as these subregions and these activities generate signals on different frequencies:

- __Motor Activity__: Often generates rhythmic activity in the beta range (13-30 Hz) during rest and suppresses this during movement.
- __Cognitive Tasks__: Associated with changes in theta (4-8 Hz) and gamma (30-100 Hz) bands.
- __Speech Production__: Activates distinct patterns in Broca's area, often involving high-frequency oscillations.

#### Electrode positioning

There are four electrodes positioning for the frontal lobe:

1. F3  - collects signals for logical reasoning, attention, decesion making, language processing & positive emotions
2. F4  - emotional regulation, negative emotions, intuitive thinking 
3. F7
4. F8
